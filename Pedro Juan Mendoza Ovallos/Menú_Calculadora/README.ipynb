## Descripción
Este proyecto es una **calculadora simple** implementada utilizando **Tkinter**, la biblioteca estándar de Python para crear interfaces gráficas. Permite realizar operaciones básicas como suma, resta, multiplicación, división, calcular el valor absoluto, el mínimo común múltiplo (MCM) y el máximo común divisor (MCD).

## Uso
1. Ejecuta el script Python:
    ```bash
    python calculadora_simple.py
    ```
2. Se abrirá una ventana con campos de entrada para los números y botones para las operaciones.
3. Ingresa los números en los campos correspondientes y selecciona la operación deseada.
4. El resultado se mostrará en el campo de resultado.

## Estructura del código
El código se compone de un solo archivo: `calculadora_simple.py`. A continuación, se detalla su estructura interna:

### Funciones Principales
- **fnSumar**: Suma dos números ingresados.
- **fnResta**: Resta dos números ingresados.
- **fnMulti**: Multiplica dos números ingresados.
- **fnDivi**: Divide dos números ingresados.
- **fnValorA**: Calcula el valor absoluto del resultado.
- **fnMCM**: Calcula el mínimo común múltiplo de dos números.
- **fnMCD**: Calcula el máximo común divisor de dos números.
- **fnSalir**: Cierra la ventana de la calculadora.

### Interfaz Gráfica
El juego utiliza **Tkinter** para crear la interfaz gráfica. Los componentes principales incluyen:
- **Etiquetas (Labels)**: Muestran los textos descriptivos y el resultado de las operaciones.
- **Entradas (Entry)**: Campos donde el usuario ingresa los números.
- **Botones (Buttons)**: Permiten al usuario realizar operaciones matemáticas.

### Ventana Principal
La ventana principal (`vent`) se configura con un título, y todos los elementos se colocan en la ventana utilizando el método `place()` para una disposición precisa.
