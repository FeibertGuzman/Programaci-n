## Descripción
Este proyecto es una aplicación gráfica simple creada con **Tkinter** en Python. Permite convertir texto ingresado en diferentes formatos: todo en mayúsculas, todo en minúsculas, solo la primera letra en mayúscula, o capitalizar la primera letra de cada palabra.

## Uso
1. Ejecuta el script Python:
    ```bash
    python convertidor_texto.py
    ```
2. Se abrirá una ventana con un campo de texto.
3. Escribe el texto que deseas convertir.
4. Usa los botones para convertir el texto:
    - **Convertir a Mayúsculas**: Convierte todo el texto a mayúsculas.
    - **Convertir a Minúsculas**: Convierte todo el texto a minúsculas.
    - **Convertir la Primera letra en Mayúscula**: Pone la primera letra del texto en mayúscula.
    - **Capitalizar Primeras Letras de Cada Palabra**: Convierte a mayúscula la primera letra de cada palabra.

## Estructura del código
El código se compone de un solo archivo principal: `convertidor_texto.py`, que implementa toda la funcionalidad del programa. A continuación, se detalla la estructura interna:

### Funciones de Conversión
El código cuenta con varias funciones que manejan las transformaciones de texto:
- **convertir_mayusculas**: Convierte todo el texto ingresado a mayúsculas.
- **convertir_minusculas**: Convierte todo el texto ingresado a minúsculas.
- **convertir_primera_mayuscula**: Convierte solo la primera letra del texto ingresado a mayúscula.
- **capitalizar_palabras**: Convierte la primera letra de cada palabra del texto ingresado a mayúscula.

Cada una de estas funciones obtiene el texto desde el campo de entrada y lo modifica directamente, actualizando el campo con el resultado convertido. Si no hay texto en el campo, se muestra una advertencia emergente utilizando `messagebox`.

### Interfaz Gráfica
El código utiliza **Tkinter** para crear la interfaz gráfica. La ventana principal (`ventana`) incluye los siguientes componentes:

#### Campo de Entrada
- **entrada**: Un `Entry` donde el usuario puede ingresar el texto a convertir. Se sitúa en la parte superior de la ventana.

#### Botones de Conversión
- **boton_mayusculas**: Ejecuta la función `convertir_mayusculas` para convertir el texto a mayúsculas.
- **boton_minusculas**: Ejecuta la función `convertir_minusculas` para convertir el texto a minúsculas.
- **boton_primera_mayuscula**: Ejecuta la función `convertir_primera_mayuscula` para capitalizar la primera letra del texto.
- **boton_capitalizar_palabras**: Ejecuta la función `capitalizar_palabras` para capitalizar la primera letra de cada palabra.

Cada botón se coloca de forma vertical en la ventana principal usando el método `pack()`.

### Ventana Principal
La ventana principal (`ventana`) está configurada con un título y tamaño fijo. El ciclo principal de la interfaz gráfica (`ventana.mainloop()`) asegura que la ventana permanezca activa, esperando interacción del usuario.
