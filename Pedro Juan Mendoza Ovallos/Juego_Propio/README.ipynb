## Descripción
"Adivina el Número" es un sencillo juego desarrollado con **Tkinter**, donde el jugador tiene que adivinar un número aleatorio entre 1 y 100 en un máximo de 10 intentos. Después de cada intento, se proporciona retroalimentación sobre si el número adivinado es demasiado bajo, demasiado alto o correcto.

## Uso
1. Ejecuta el script Python:
    ```bash
    python adivina_el_numero.py
    ```
2. Se abrirá una ventana que te pedirá adivinar un número entre 1 y 100.
3. Ingresa tu número en el campo de texto y presiona el botón "Adivinar".
4. Recibirás retroalimentación sobre tu intento y se te indicará cuántos intentos quedan.
5. Si adivinas el número correctamente o se te acaban los intentos, el juego terminará.

## Estructura del código
El código se compone de un solo archivo: `adivina_el_numero.py`. A continuación, se detalla su estructura interna:

### Funciones Principales
- **actualizar_resultado(resultado)**: Actualiza la interfaz con el resultado del intento y muestra los intentos restantes.
- **adivinar()**: Maneja la lógica para verificar la adivinanza del usuario y proporciona retroalimentación.

### Interfaz Gráfica
La interfaz se crea utilizando **Tkinter**, e incluye:
- **Etiquetas (Labels)**: Para mostrar instrucciones, resultados y el número de intentos restantes.
- **Entrada (Entry)**: Campo donde el usuario ingresa su número.
- **Botón (Button)**: Para enviar la adivinanza.

### Variables Globales
- **numero_a_adivinar**: Almacena el número aleatorio que el jugador debe adivinar.
- **intentos**: Contador de los intentos restantes del jugador.
