## Descripción
Esta es una **calculadora de suma** simple desarrollada utilizando **HTML**, **CSS** y **JavaScript**. Permite a los usuarios introducir dos números y calcular su suma al presionar un botón. La interfaz es sencilla y está diseñada para ser intuitiva.

---

1. Clona el repositorio en tu máquina local:

2. Navega a la carpeta del proyecto:
    ```bash
    cd tu-repo
    ```
3. Abre el archivo `index.html` en tu navegador preferido.

---

## Uso
1. Abre el archivo `index.html` en tu navegador.
2. Introduce dos números en los campos correspondientes.
3. Haz clic en el botón "Sumar".
4. El resultado aparecerá debajo del formulario.

---

## Estructura del código
El proyecto consta de tres archivos principales:

### 1. `index.html`
Contiene la estructura básica de la página, incluyendo el formulario para ingresar los números y un lugar para mostrar el resultado.

### 2. `script.js`
Maneja la lógica del cálculo. Contiene la función `sumaNumeros()` que se ejecuta al enviar el formulario y calcula la suma de los números ingresados.

### 3. `style.css`
Define los estilos visuales de la aplicación, incluyendo el formato de textos, colores y disposición de los elementos en la página.
