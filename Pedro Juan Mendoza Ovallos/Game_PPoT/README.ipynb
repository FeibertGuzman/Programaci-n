## Descripción
Este proyecto es un juego de "Piedra, Papel o Tijera" implementado utilizando **Tkinter**, la biblioteca estándar de Python para crear interfaces gráficas. El usuario compite contra la computadora en un total de 10 partidas y puede ver los resultados acumulados de victorias, derrotas y empates.

---

## Instalación
1. Clona el repositorio en tu máquina local:

## Uso
1. Ejecuta el script Python:
    ```bash
    python piedra_papel_tijera.py
    ```
2. Se abrirá una ventana con opciones para elegir "Piedra", "Papel" o "Tijera".
3. Selecciona una opción, y el juego mostrará qué eligió la computadora y el resultado.
4. El juego continuará hasta que se completen las 10 partidas, tras lo cual mostrará el resultado final.

## Estructura del código
El código se compone de un solo archivo: `piedra_papel_tijera.py`. A continuación, se detalla su estructura interna:

### Funciones Principales
- **jugar**: Esta función se activa cuando el usuario selecciona una opción (piedra, papel o tijera). Compara la elección del usuario con la de la computadora y determina el resultado de la partida (victoria, derrota o empate). Luego actualiza las variables de puntuación y el número de partidas restantes.
- **actualizar_resultado**: Esta función actualiza la interfaz gráfica con la elección de la computadora, el resultado de la partida, y las estadísticas de victorias, derrotas, empates y partidas restantes. Si las partidas llegan a cero, desactiva los botones de juego y muestra el resultado final.

### Interfaz Gráfica
El juego utiliza **Tkinter** para crear la interfaz gráfica. Los componentes principales incluyen:
- **Etiquetas (Labels)**: Muestran la elección de la computadora, el resultado de la partida, y las estadísticas de victorias, derrotas y empates.
- **Botones (Buttons)**: Tres botones permiten al usuario seleccionar piedra, papel o tijera para jugar.

### Ventana Principal
La ventana principal (`root`) se configura con un título, y todos los elementos se colocan verticalmente en la ventana usando el método `pack()`.
