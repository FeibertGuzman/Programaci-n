### Descripción

Este proyecto es un juego interactivo de Piedra, Papel o Tijera, implementado en Python utilizando la biblioteca Tkinter para la interfaz gráfica. El jugador puede elegir entre las tres opciones y el programa selecciona aleatoriamente una de ellas para competir. El objetivo es alcanzar 10 victorias antes de que el programa lo haga, y se lleva un registro de victorias, empates y derrotas.

### Instalación
Para ejecutar este proyecto, asegúrate de tener Python y Tkinter instalados en tu sistema. Tkinter generalmente viene incluido con la instalación estándar de Python.

### Herramientas utilizadas:

Python: Lenguaje de programación utilizado para desarrollar el juego.

Tkinter: Biblioteca estándar de Python para la creación de interfaces gráficas de usuario.

random: Módulo de Python utilizado para seleccionar aleatoriamente la opción del programa.
Uso

Al iniciar el programa, se abrirá una ventana que mostrará un campo de entrada para que el jugador elija entre Piedra, Papel o Tijera. El jugador puede hacer su selección haciendo clic en los botones correspondientes. El juego mostrará el resultado de cada ronda, así como un conteo actualizado de victorias, empates y derrotas. El juego se detendrá automáticamente una vez que el jugador alcance 10 victorias.

4. ### Uso

- Ejecutar el programa: Al ejecutar el archivo Python que contiene el código, se abrirá una ventana gráfica que presenta el juego.

- Seleccionar una opción: El jugador puede elegir entre tres botones: Piedra, Papel o Tijera. Al hacer clic en uno de ellos, se iniciará una ronda del juego.

- Ver el resultado: Después de hacer una selección, el juego mostrará el resultado de la ronda en la etiqueta correspondiente, indicando si el jugador ganó, perdió o si fue un empate. Además, se mostrará la elección de la máquina.

- Contadores: En la interfaz se visualizarán contadores que indican cuántas victorias, empates y derrotas ha tenido el jugador hasta el momento.

- Finalizar el juego: El juego continuará hasta que el jugador alcance 10 victorias. En ese momento, se mostrará un mensaje indicando que el juego ha finalizado y los botones de selección se desactivarán.

- Reiniciar el juego: Para jugar de nuevo, será necesario reiniciar el programa.

5. ### Estructura del código

**Funciones:**
jugar(opcion_usuario): Maneja la lógica del juego, determina el resultado de cada ronda y actualiza la interfaz.

actualizar_contadores(): Actualiza los contadores de victorias, empates y derrotas en la interfaz gráfica.

desactivar_botones(): Desactiva los botones de selección una vez que se alcanza el objetivo de 10 victorias.

**lementos de la interfaz:**
Ventana principal (ventana).

Etiquetas (Label) para mostrar el resultado y contadores.

Botones (Button) para las opciones de juego (Piedra, Papel, Tijera).
