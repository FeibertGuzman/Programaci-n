## Descripción
Este proyecto es una aplicación de escritorio simple desarrollada con Python que permite a los usuarios convertir texto entre diferentes formatos: mayúsculas, minúsculas, la primera letra en mayúscula y capitalizar las primeras letras de cada palabra. La aplicación utiliza la biblioteca Tkinter para crear una interfaz gráfica amigable.

## Instalación
Para ejecutar este proyecto, necesitas tener Python instalado en tu sistema. Sigue estos pasos para instalar y ejecutar la aplicación:

1. Clona este repositorio en tu máquina local:
   ```bash
https://github.com/FeibertGuzman/Programaci-n/tree/4e659627ed305171f279af8b403a06f5933cd583/Miguel%20%C3%81ngel%20Ram%C3%ADrez%20Corredor/Calculador%20de%20MM   ```
   
2. Navega al directorio del proyecto:
   ```bash
   cd Calculadora_Mayusculas_Minusculas
   ```

3. Asegúrate de que tienes Tkinter instalado. Tkinter viene preinstalado con Python en la mayoría de las distribuciones, pero si no lo tienes, puedes instalarlo utilizando el siguiente comando:
   ```bash
   sudo apt-get install python3-tk
   ```

4. Ejecuta la aplicación:
   ```bash
   python nombre_del_archivo.py
   ```

## Herramientas Utilizadas
- **Python**: Lenguaje de programación utilizado para desarrollar la aplicación.
- **Tkinter**: Biblioteca estándar de Python para crear interfaces gráficas de usuario.

## Uso
1. Al abrir la aplicación, verás un campo de entrada donde puedes escribir o pegar el texto que deseas convertir.
2. Utiliza los botones disponibles para realizar las siguientes conversiones:
   - **Convertir a Mayúsculas**: Convierte todo el texto a mayúsculas.
   - **Convertir a Minúsculas**: Convierte todo el texto a minúsculas.
   - **Convertir la Primera Letra en Mayúscula**: Convierte solo la primera letra del texto a mayúscula.
   - **Capitalizar Primeras Letras de Cada Palabra**: Capitaliza la primera letra de cada palabra en el texto.

## Estructura del Código
- **Funciones**: Cada botón tiene una función asociada que se encarga de la conversión del texto:
  - `convertir_mayusculas()`: Convierte el texto a mayúsculas.
  - `convertir_minusculas()`: Convierte el texto a minúsculas.
  - `convertir_primera_mayuscula()`: Convierte la primera letra del texto a mayúscula.
  - `capitalizar_primeras_letras()`: Capitaliza la primera letra de cada palabra.

- **Interfaz de Usuario**: Se crea una ventana principal con un campo de entrada y botones para cada conversión, como se puede observar a continuación: [Interfaz gráfica](./Interfaz.png)
