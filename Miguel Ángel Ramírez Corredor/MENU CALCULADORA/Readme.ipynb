# Calculadora Tkinter

Este proyecto es una calculadora básica desarrollada en Python utilizando la biblioteca Tkinter. La aplicación proporciona una interfaz gráfica para realizar operaciones matemáticas comunes.

## Características

- **Ventana Principal y Menú:**
  - La ventana principal contiene un menú con las siguientes opciones matemáticas:
    - Suma
    - Resta
    - Multiplicación
    - División
    - Valor Absoluto
    - Mínimo Común Múltiplo (MCM)
    - Máximo Común Divisor (MCD)

- **Operaciones Matemáticas:**
  - Cada opción en el menú realiza la operación matemática correspondiente y muestra el resultado en un campo de texto en la interfaz.
