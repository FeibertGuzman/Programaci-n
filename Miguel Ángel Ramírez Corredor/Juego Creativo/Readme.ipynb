## Descripción
Este proyecto es una aplicación de juego de memoria, donde los jugadores deben unir pares de colores. El juego se basa en una interfaz gráfica creada con Python y Tkinter. La aplicación mezcla colores y permite a los jugadores seleccionar botones para descubrir pares.

## Instalación
Para ejecutar este proyecto, necesitas tener Python instalado en tu sistema. Sigue estos pasos para instalar y ejecutar la aplicación:

1. **Clona este repositorio en tu máquina local:**
   ```bash
   git clone https://github.com/FeibertGuzman/Programaci-n.git

2. **Navega al directorio del proyecto**

Juego.py

3. **Instalar Tkinter** 
Tkinter generalmente viene incluido con Python en la mayoría de las distribuciones, pero si no lo tienes, puedes instalarlo utilizando el siguiente comando:
sudo apt-get install python3-tk

4. **Ejecuta la aplicación**
python Juego.py

## Herramientas Utilizadas

- Python: Lenguaje de programación utilizado para desarrollar la aplicación.

- Tkinter: Biblioteca estándar de Python para crear interfaces gráficas de usuario.

## Uso

Al abrir la aplicación, verás una cuadrícula de botones representando colores ocultos.

Haz clic en dos botones para intentar encontrar pares de colores. Si seleccionas dos colores que coinciden, se mantendrán visibles. Si no coinciden, se ocultarán nuevamente.

El juego termina cuando se han encontrado todos los pares.

## Estructura del Código

- Clase JuegoMemoria: Contiene la lógica del juego, incluyendo la inicialización de la ventana, creación de botones, manejo de selecciones y verificación de pares.

- Métodos:
. seleccionar(i, j): Maneja la lógica de selección de los botones.

. verificar_pareja(): Verifica si los colores seleccionados son un par.

. mostrar_mensaje(mensaje): Muestra un mensaje al final del juego.
