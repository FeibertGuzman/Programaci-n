## Convertidor de Texto
Este proyecto es una aplicación gráfica simple construida en Python utilizando la librería Tkinter. La aplicación permite realizar diferentes conversiones de texto como convertir a mayúsculas, minúsculas, capitalizar la primera letra de cada palabra, entre otras. Es ideal para mostrar el uso de elementos básicos de Tkinter como botones, entradas de texto y etiquetas.

## Requisitos
Python 3.x instalado en el sistema.
No es necesario instalar dependencias adicionales, ya que Tkinter es parte del paquete estándar de Python.

## Cómo ejecutar el proyecto
Descargue o clone este repositorio.

Abra una terminal o consola de comandos.

Navegue a la carpeta donde se encuentra el archivo convertidor_de_texto.py.

Ejecute el archivo utilizando Python con el siguiente comando:

bash

Copiar código
python convertidor_de_texto.py

## Funcionalidades
Entrada de texto
La aplicación cuenta con un cuadro de entrada donde puedes escribir el texto que deseas convertir. Este cuadro de entrada acepta cadenas de texto y tiene un ancho de 50 caracteres.

## Botones de conversión
La aplicación ofrece cuatro botones, cada uno con una función diferente:

Convertir a Mayúsculas: Convierte todo el texto ingresado a mayúsculas.

Convertir a Minúsculas: Convierte todo el texto ingresado a minúsculas.

Convertir la Primera Letra en Mayúscula: Convierte solo la primera letra del texto a mayúscula, el resto en minúsculas.

Capitalizar Primeras Letras de Cada Palabra: Convierte la primera letra de cada palabra en mayúscula.

## Resultado
El resultado de la conversión se muestra en una etiqueta debajo de los botones, actualizándose dinámicamente según la operación seleccionada.

## Estructura del Código

Funciones de conversión:

convertir_mayusculas(): Toma el texto de la entrada y lo convierte a mayúsculas.

convertir_minusculas(): Toma el texto de la entrada y lo convierte a minúsculas.

convertir_primera_letra_mayuscula(): Convierte solo la primera letra del texto ingresado a mayúscula.

capitalizar_primera_letra_cada_palabra(): Capitaliza la primera letra de cada palabra en el texto.

## Interfaz gráfica:

Se crean botones y etiquetas usando Tkinter para permitir la interacción del usuario.

La función StringVar() es utilizada para actualizar dinámicamente el texto mostrado en la etiqueta de resultado.

## Estructura de la Interfaz

Entrada de Texto: Cuadro donde el usuario ingresa el texto a convertir.

Botones de Conversión: Cuatro botones que realizan diferentes conversiones de texto.

Resultado: Etiqueta que muestra el resultado de la conversión en tiempo real.

## Ejemplo de uso
Escribe un texto en la entrada, como: hola mundo.

Presiona el botón de "Convertir a Mayúsculas". El resultado será HOLA MUNDO.

Presiona el botón de "Convertir a Minúsculas". El resultado será hola mundo.

Prueba con los otros botones para realizar diferentes conversiones.
